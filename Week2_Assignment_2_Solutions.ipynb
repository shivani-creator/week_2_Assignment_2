## Machine Learning: Concepts & Algorithms

## Q1. Loan approval prediction — classification or regression?

This is a **classification problem**, since the output is a discrete category (approved or rejected).

Among the three algorithms, a **Decision Tree** is most suitable. KNN can also classify, but Linear Regression predicts continuous values and is unsuitable for a yes/no decision.

## Q2. Feature vs. Label

A **feature** is an input variable used to make a prediction; a **label** is the output we want to predict.

**Example:** predicting a house price — the features are area, number of bedrooms, and location, while the label is the price.

## Q3. How a Decision Tree reaches a prediction

A Decision Tree splits the data using a series of yes/no questions on features. Starting at the root, it selects the feature that best separates the classes (using a criterion such as **Gini impurity** or **information gain**), branches accordingly, and repeats at each node. The customer's feature values follow a path down the branches until reaching a **leaf node**, whose majority class ("will purchase") becomes the prediction.

## Q4. Why is KNN supervised? Can it classify without labels?

KNN is **supervised** because it requires labeled training data — it predicts a new point's label from the known labels of its nearest neighbors. Without labels, KNN cannot perform classification, since there are no class labels to vote on. It can only be adapted for unsupervised tasks such as clustering or anomaly detection.

## Q5. KNN prediction for K = 5

**Neighbors:** Yes, Yes, No, Yes, No → 3 "Yes" and 2 "No".

KNN uses **majority voting**, so the prediction is **Yes**, because the most frequent class among the 5 neighbors wins.

```python
# Q5 demonstration: KNN majority voting for K = 5
from collections import Counter

neighbors = ["Yes", "Yes", "No", "Yes", "No"]
prediction = Counter(neighbors).most_common(1)[0][0]

print("Neighbor labels:", neighbors)
print("Vote counts:    ", dict(Counter(neighbors)))
print("Prediction:     ", prediction)
```

## Q6. Equation of Linear Regression: $y = mx + b$

- $x$ — input / independent variable (feature)
- $y$ — predicted output / dependent variable
- $m$ — slope (change in $y$ per unit change in $x$)
- $b$ — intercept (value of $y$ when $x = 0$)

## Q7. Correlation

Correlation measures the **strength and direction** of a linear relationship between two variables.

- **+1** — perfect positive correlation (both increase together)
- **0** — no linear relationship
- **−1** — perfect negative correlation (one increases as the other decreases)

## Q8. Disease present, predicted healthy

The model predicted "Healthy" when the disease was actually present — this is a **False Negative**.

It is dangerous because a sick patient is told they are healthy, so they miss treatment, allowing the disease to progress or spread.

## Q9. Confusion Matrix vs. Correlation Matrix

A **Confusion Matrix** summarizes a classification model's performance by counting True Positives, True Negatives, False Positives, and False Negatives, showing where predictions match or differ from actual labels.

A **Correlation Matrix** is different: it shows correlation coefficients between pairs of variables (features), measuring relationships within the data itself — not model performance.

## Q10. Comparison of algorithms

- **Decision Tree** — splits data into branches using feature-based questions to reach a decision.
- **KNN** — classifies a point based on the majority label of its K closest neighbors.
- **SVM** — finds the optimal hyperplane that maximally separates the classes.
- **Linear Regression** — fits a straight line to model the relationship between inputs and a continuous output.
- **Neural Network** — layers of interconnected neurons that learn complex patterns through weighted connections.
